```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
company_stock = pd.read_csv ('stock_data.csv')
company_stock
```

```python
company_stock.plot (x = 'Date', y = 'TWTR', label = 'Twitter Stock Price', figsize = (15,15), linewidth = (1), color = 'white')
plt.title ('Twitter Stock')
plt.legend (loc = 'upper left')
plt.xlabel ('Date')
plt.ylabel ('Price [$]')
plt.grid()
```

```python
daily_return = pd.read_csv ('stocks_daily_returns.csv')
daily_return
```

```python
daily_return.plot (x = 'Date', y = 'TWTR', label = 'Stocks Daily Returns', figsize = (15,15), linewidth = (1), color = 'white')
plt.grid()
plt.title ('Twitter Stocks Daily  Return')
plt.legend (loc = 'upper left')
plt.xlabel ('Date')
plt.ylabel ('Price [$]')
```

```python
daily_return.plot (x = 'Date', y = 'NFLX', label = 'Stocks Daily Returns', figsize = (15,15), linewidth = (1), color = 'white')
plt.grid()
plt.title ('Netflix Stocks Daily  Return')
plt.legend (loc = 'upper left')
plt.xlabel ('Date')
plt.ylabel ('Price [$]')
```
